**Find out correlation between variable in the given dataset correlation between different variables.**

```python
label_encoder = LabelEncoder()
for column in df.select_dtypes(include=['object']).columns:
    df[column] = label_encoder.fit_transform(df[column])
```

```python
correlation_matrix = df.corr()
print(correlation_matrix)
```

```python
plt.figure(figsize=(10, 10))
sns.heatmap(correlation_matrix, cbar=True, annot=True, cmap='Blues')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

**plot horizontal bar chart to analyze the reduced days of activity due to illness base on gender**

```python
db= df.groupby(['gender'])['reduced'].sum().to_frame().reset_index()
#creating the bar chart
plt.barh(db['gender'],db['reduced'],color =['cornflowerblue','lightseagreen'])
#Adding the  aesthetics
plt.title('Bar chart')
plt.xlabel('reduced activity')
plt.ylabel('gender')
plt.show()
```
